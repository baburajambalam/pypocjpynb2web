```python
def getInflationMultiplicationFactor():
    rate_of_inflation = (w_inflation_rate_savings_f_nr.value/100)
    return (1 + rate_of_inflation) 

#print("get inflation multiplication factor", getInflationMultiplicationFactor())
```

```python
def getEarningMultiplicationFactor():
    rate_of_earning = (w_earn_percent_on_investment_f_nr.value/100)
    return (rate_of_earning)

#print("get earning multiplication factor", getEarningMultiplicationFactor())
```

```python
def getYearlyRegularIncome(debug, age):
    #income for 12 months
    income_before_retirement = w_income_before_retirement_nr.value * 12 * (age<=w_retirement_age_nr.value) 
    income_after_retirement = w_monthly_income_afer_retiring_nr.value * 12 * (age>w_retirement_age_nr.value)
    income_from_social_security = w_social_security_income_nr.value * 12 * (age>w_retirement_age_nr.value)
    if debug:
        print({"income_before_retirement":income_before_retirement ,"income_after_retirement":income_after_retirement ,
               "income_from_social_security": income_from_social_security})
    return  income_before_retirement + income_after_retirement + income_from_social_security        
```

```python
def getIncomeForYear(debug=False, investment_amt=0, age=35, annual_expense=10000):   
    new_age = age + 1     
    investment_amt_local = int(investment_amt * (investment_amt>0))
    
    new_annual_expense = int(annual_expense *  getInflationMultiplicationFactor())
    new_annual_expense = new_annual_expense * (new_annual_expense>0)
    #print("annual_expense", annual_expense, " new_annual_expense", new_annual_expense)    
        
    investment_earning = int(investment_amt_local * getEarningMultiplicationFactor())
    #print("getEarningMultiplicationFactor()", getEarningMultiplicationFactor()) 
    
    new_investment_amt =  int(investment_amt_local + getYearlyRegularIncome(debug, new_age) 
                              + investment_earning - new_annual_expense)   
    financial_aid_required = abs(new_investment_amt) * (new_investment_amt<=0)
    new_investment_amt = new_investment_amt * (new_investment_amt>0)
    #print("new_investment_amt-B", new_investment_amt)
    
    inData = {"investment_amt":investment_amt,"age": age, "annual_expense": annual_expense}
    outData = {"new_investment_amt":new_investment_amt,"new_age": new_age, 
             "new_annual_expense": new_annual_expense, "financial_aid_required":financial_aid_required }
    
    if debug:        
        print({"new_annual_expense":new_annual_expense,"investment_earning":investment_earning,
               "new_investment_amt":new_investment_amt, "financial_aid_required":financial_aid_required })
        print(inData)
        print (outData, end="\n-----\n")
        
    return (outData)

#print( getIncomeForYear(debug,0, 70, 40003))
```
